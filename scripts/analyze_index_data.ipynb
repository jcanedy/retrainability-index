```python
import polars as pl
import seaborn as sns
```

```python
DATA_FOLDER = "../data/processed"
data_lazy = pl.scan_parquet(f"{DATA_FOLDER}/index_data.parquet")
```

```python
data_lazy
```

```python
data = data_lazy.collect()
```

```python
(
    data
    .filter(
        pl.col("has_occupational_outcomes"),
    )
    .group_by(
        pl.col("is_wages_mean_increased"),
    )
    .len()
)
```

```python
participation_durations = (
    data.select(
        (pl.col("exit_date") - pl.col("entry_date")).alias("participation_duration")
    )
    .filter(
        (pl.col("participation_duration") > 0) & ~(pl.col("participation_duration").is_null())
    )
)
```

```python
sns.histplot(data=participation_durations.to_numpy().reshape(1, -1)[0])
```
